```python
# # Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python. It provides a wide variety of plots and charts for data visualization purposes. Matplotlib is extensively used in data analysis, scientific computing, and machine learning due to its flexibility and ease of use.

# The Pyplot module in Matplotlib provides a MATLAB-like interface for creating plots and visualizations. Here are five types of plots that can be created using the Pyplot module:

# 1 Line Plot: A basic plot that displays data points connected by straight line segments.
# 2 Scatter Plot: A plot that displays individual data points as markers, useful for visualizing relationships between two variables.
# 3 Histogram: A plot that represents the distribution of a single variable by dividing the data into bins and displaying the frequency of occurrences in each bin.
# 4 Bar Plot: A plot that represents categorical data with rectangular bars, where the length of each bar corresponds to the value of the data.
# 5 Pie Chart: A circular statistical graphic that is divided into slices to illustrate numerical proportion.
```

```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
# plot a scatter plot.
# Scatter Plot: A plot that displays individual data points as markers, useful for visualizing relationships between two variables.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# is used to create a grid of subplots within a single figure. This function is particularly useful when you want to display multiple plots or visualizations together in a single figure,
# organized in a grid layout.

# For line 1: 
x1= np.array([0, 1, 2, 3, 4, 5]) 
y1= np.array([0, 100, 200, 300, 400, 500])
# For line 2: 
x2= np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
# For line 3: 
x3= np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
# For line 4: 
x4= np.array([0, 1, 2, 3, 4, 5])
y4= np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x1,y1)
plt.subplot(2,2,2)
plt.plot(x2,y2)
plt.subplot(2,2,3)
plt.plot(x3,y3)
plt.subplot(2,2,4)
plt.plot(x4,y4)
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

# 
# # # # A bar plot, also known as a bar chart, is a type of chart that represents categorical data with rectangular bars.
# # # The length or height of each bar corresponds to the value it represents.
# # Bar plots are commonly used to visualize and compare the frequency, count, or proportion of different categories or groups within a dataset.

# bar plot used 
# comparison 
# categorical data visualisations
# presenting of frequencies
# readibilities

import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# 
# A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset through five summary statistics: minimum, first quartile (Q1), median (second quartile, Q2), third quartile (Q3), and maximum.
# It provides a visual summary of the central tendency, dispersion, and skewness of the data, as well as identifying outliers
# boxplot used for 
# dummary  of distribution
# identification of outliers 
# comparison of groups
# robustness of skewed data

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
data=[box1,box2]
plt.boxplot(data,labels=['box1','box2'])
```
